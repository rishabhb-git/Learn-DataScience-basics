## Groupings

- via `.groupby()`

Internally, how Pandas perfrom Grouping ?
1. When `.groupby([col1, col2])` is called on a dataframe, for a column, firstly Pandas creats smaller "groups" w.r.t. those columns (i.e. col1, col2, etc.).


2. Secondly, Pandas "apply" an aggregate function like - sum, min, max, count, etc.


3. Thirdly and finally, Pandas would "combine" the aggregated results.


<h3 style="text-align:center;">GROUPBY -> APPLY of aggregate functions -> Combine</h3>

__Scneario 11__: find maximum budget spend by each director.

```python
## step 1.

df.groupby(["director_name"])
```

```python
## step 2. and 3.

df.groupby(["director_name"])["budget_Mdollar"].max()
```

<br>__Scenario 13__: find the "Most Productive Director" - the director who produced most number of movies per year.

```python
df.groupby(["director_name"])["year"].min().head()

## when did a Director released his first movie. Or, in which year did a Director start his career.
```

```python
df.groupby(["director_name"])["year"].max().head()
```

For each `director_name`, if we subtract his min-year value from his max-year value, we would get the career span of each director. Now, for each director, his total movies divided by his career span would give total movies produced per year.

The maximum of this value would correspond to "Most Productive Director".

Let's perform a groupby on `director_name` and perform aggregations on `title` and `year` columns, where we would need aggregate-count of `title` column, and aggregate-min & aggregate-max of `year` column.

This process is essentially : __GROUPBY -> APPLY of agg functions -> COMBINE__

```python
df_agg = df.groupby(["director_name"])[["title" , "year"]].aggregate({"title" : "count",
                                                             "year" : ["min", "max"]
                                                            })

df_agg
```

```python
df_agg["year"]["max"].head()

## doing this gives same Series like we got above
```

```python
df_agg.columns
```

Now, it's indeed cumbersome to use and understand `df_agg` , because its a Multilevel-Indexed dataframe.

1. So let's convert its columns into Linear-Indexed.


2. After step 1, reset the index and make `director_name` as a normal column.

```python
df_agg.columns = ["_".join(col) for col in df_agg.columns]      ## step 1

df_agg.reset_index(inplace=True)                                ## step 2

df_agg
```

```python
## create new column of "active_years"

df_agg["active_years"] = df_agg["year_max"] - df_agg["year_min"]

df_agg.head()
```

<br>Now,

Productivity of a Director = Movies made by him per year = __title_count Ã· active_years__

```python
df_agg["productivity_of_director"] = df_agg["title_count"] / df_agg["active_years"]

df_agg
```

```python
df_agg.sort_values(["productivity_of_director"], ascending=False, inplace=True)

# df_agg.reset_index(inplace=True)
### no need to do reset_index, because this will convert the existing implicit indices into a new column namely-"index". 

df_agg.head()
```

Hence, these are our top 5 Most Productive Directors!!

<br>__Scenario 16__: Find the details of all such movies, that were made by a "High Budget Director".

This is similar to __Scenario 15__, and after finding out high-budget-directors, we can find all movies of every high-budget-director, via join or merge. But, that would be lot of operations and lengthy approach.

We can use __Group-based-Filtering__ to achieve this.

- `.filter()`

Approach :- Create groups (via `groupby`) of each director. Pass each group in a filter-based function that returns `True` if the director's maximum budgeted movie has been above $ 1M, otherwise `False`. Certainly, the boolean value corresponds to the director_name; `True` means such a director is "High Budget Director".

```python
df.groupby(["director_name"]).filter(lambda x: x["budget_Mdollar"].max() >= 100)

## lambda function says, if any director's maximum budget (GROUP -> APPLY) 
```

<br>__Scenario 17__:

Which movies were "Expensive" (or, "Over Budget") an which were "Inexpensive".

<code>
    budget_minus_meanBudget = budget - budget.mean()
    if budget_minus_meanBudget > 0:
        "Expensive or Over Budget"
    else:
        "Inexpensive"
</code>

```python
df_transform = df.copy()

df_transform
```

```python
def sub_mean(x):
    return x["budget_Mdollar"] - x["budget_Mdollar"].mean()
    

df_transform["budget_minus_meanBudget"] = df_transform.groupby(["director_name"]).transform(sub_mean)

df_transform
```

> this results in KeyError because, there is a limitation of `transform` that it allows __only 1 column__ to be used with it. Since we are operating on "budget_Mdollar" column, so after doing `groupby()` we'll have to fetch out "budget_Mdollar" and then proceed for the `transform`.

â›” Generally, `transform` is not preferred because it can only be used on 1 column, at 1 time.<br>
Whereas with `apply` multiple columns can be in single time, so it's mostly preferred.

Above usage is wrong, Correct way to use `transform` is this :

```python
def sub_mean(x):
    return round(x - x.mean(), 2)


df_transform["budget_minus_meanBudget"] = df_transform.groupby(["director_name"])["budget_Mdollar"].transform(sub_mean)
```

<br>__Scenario 18__: Which movie was a "risky" investment. Also discussed [here in Pandas-3 file](./Pandas-3.ipynb#Scenario-9:-Finding-risky-movie,-using-apply-function)

syntax : `apply`

This can be found via the logic that, if average revenue of the director is more than the movie's budget, then the movie can be said as "risky". To do so, we can groupby the dataframe by the director's name, and using `apply` function, find the required data.

Means, for a Director,
- `avg(revenue)` of all his movies > `budget` of 1 movie => "Risky" movie

```python
def calculate_risky(x):
    x["risky"] = (x["budget_Mdollar"] - x["revenue_Mdollar"].mean()) > 0     ## RHS generates either True or False value
    return x


df_risky = df.groupby("director_name").apply(calculate_risky)

df_risky
```

<br>Let's see which movies were actually "Risky" :

```python
df_risky[df_risky["risky"] == True]
```
