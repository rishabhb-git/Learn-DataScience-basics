## B. Making figures/plots using NumPy arrays
One of things that matter is figuring out what kind of plot to use to get a better feel for the data. 

Matplotlib visualizations are built off NumPy arrays. And most common types of plots available in matplotlib are :

- line
- scatter
- bar
- hist
- `subplots()`

```python
import numpy as np
```

#### 1. Line
Line is the default type of visualization in Matplotlib.<br>**The default, unless specified otherwise, plots will start out as "line".**

```python
# Create an array
x = np.linspace(1, 10, 100)

x
```

```python
# See below, the default plot will be "line"

fig, ax = plt.subplots()
ax.plot(x, x**2);
```

#### 2. Scatter

```python
fig, ax = plt.subplots()
ax.scatter(x,np.sin(x))
```

```python
fig, ax = plt.subplots()
ax.plot(x, np.log(x));
```

#### 3. Bar
- Vertical
- Horizontal

```python
# You can make plots from a dictionary

nut_butter_prices = {"Almond butter": 18,
                     "Peanut butter": 15,
                     "Cashew butter": 25}
fig, ax = plt.subplots()

ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())

ax.set(title="Krishna's Bakery Store", ylabel="Price ($)")
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))    # .barh() for horizontal bar graph
```

#### 4. Histogram (hist)

```python
# Make some data from a normal distribution

x = np.random.randn(500) # pulls data from a normal distribution

fig, ax = plt.subplots()
ax.hist(x);
```

```python
x = np.random.random(500) # random data from random distribution

fig, ax = plt.subplots()
ax.hist(x);
```

#### 5. Sublots
Multiple plots on one figure https://matplotlib.org/3.1.1/gallery/recipes/create_subplots.html

**Method 1 :** Create multiple subplots by defining a tuple-of-tuples of all axes; ax1, ax2, ax3 and ax4.

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, 
                                             ncols=2, 
                                             figsize=(10, 5))

# Plot data to each axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

**Method 2 :** Create multiple subplots, by accessing the different axes via their index.

```python
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))

# Index to plot data
ax[0, 0].plot(x, x/2);                                                  # (x,y) beahive like, y = x/2
ax[0, 1].scatter(np.random.random(10), np.random.random(10));           # (x,y) are some random datapoints on xy-plane
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());     # (x,y) denote the sales of butter
ax[1, 1].hist(np.random.randn(100));
```

## D. Plotting with pandas using the OO method
For more complicated plots, you'll want to use the OO method.

* When plotting something quickly, it's okay to use the `pyplot`.
* When plotting something more advanced, use the OO method.

```python
import matplotlib.pyplot as plt
plt.style.use('default')
```

Perform data analysis on patients over 50 years of age.

```python
over_50 = heartDisease[heartDisease["age"] > 50]
over_50.shape                                      # 208 patients are there
```

The column `"target"` is made up of 2 kinds of entries i.e. {0, 1}.

```python
set(heartDisease['target'])
```

**Method 1 :** _normal way of doing; without OO method._

```python
over_50.plot(kind='scatter',
             x='age',            # 'age' column on x-axis
             y='chol',           # 'chol' column on y-axis
             c='target',         # 'target' column - changes the color of dots based on the number in target column
             figsize=(10, 6));
```

**Method 2 :** _with OO method (__with pyplot method__)_ 

```python
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
             x="age",
             y="chol",
             c='target',
             ax=ax);

ax.set_xlim([45, 100]);                   # adjusts the x-axis's stretch from 45 to 100 points. 
```

**Method 3 :**<br>

Creating a bit more complicated plot ___(implementing the OO method from scratch) -___

```python
# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(over_50["age"], 
                     over_50["chol"], 
                     c=over_50["target"]         # "c" is for color i.e. portraying the dot (datapoint) with a color
                    )

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age (years)",
       ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");
```

>***SIDE NOTE :***<br><br>**How `ax.legend(*scatter.legend_elements(), title="Target")` works :**<br><br>Firstly, the `.legend_elements()` will operate on `scatter` variable by querying the color parameter `c` in `ax.scatter()`. It will return the set of colors used to display/color the datapoints. If the column has only 2 types of values (like in the case of "target" column) then, the datapoints which have `target=1` will be of one color and `target=0` will be of another color.<br><br>Have a look below, if the `thal` column is used how the plot would look.<br><br>**That's how datapoints get their correspoinding color.**

```python
# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(over_50["age"], 
                     over_50["chol"], 
                     c=over_50["thal"]            # color will come from 'thal' column 
                    )

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age (years)",
       ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");
```

***To insert a horizontal line going across with the mean of `heartDisease["chol"]` ?***

```python
# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(over_50["age"], 
                     over_50["chol"], 
                     c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age (years)",
       ylabel="Cholesterol");
ax.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax.axhline(over_50["chol"].mean(),
           linestyle="--");
```

### Adding another plot to existing styled plot, using OO method

```python
# Setup plot (2 rows, 1 column)
fig, (ax0, ax1) = plt.subplots(nrows=2, # 2 rows
                               ncols=1, 
                               sharex=True, 
                               figsize=(12, 10))

# Add data for ax0
scatter = ax0.scatter(over_50["age"], 
                      over_50["chol"], 
                      c=over_50["target"])
# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")
ax0.legend(*scatter.legend_elements(), title="Target")

# Setup a mean line
ax0.axhline(y=over_50["chol"].mean(), 
            color='b', 
            linestyle='--', 
            label="Average")


# Add data for ax1
scatter = ax1.scatter(over_50["age"], 
                      over_50["thalach"], 
                      c=over_50["target"])

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate Levels",
        xlabel="Age",
        ylabel="Max Heart Rate")
ax1.legend(*scatter.legend_elements(), title="Target")

# Setup a mean line
ax1.axhline(y=over_50["thalach"].mean(), 
            color='b', 
            linestyle='--', 
            label="Average")

# Title the figure
fig.suptitle('Heart Disease Analysis', fontsize=18, fontweight='bold');
```

###  Customize limitations of `xlim` & `ylim`

_Before the change of `xlim` and `ylim`_

```python
# Axis 0, 1 (row 0, column 1)
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 15))

scatter = ax0.scatter(over_50["age"], 
                      over_50["chol"], 
                      c=over_50["target"],
                      cmap='winter')

ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol",
        xlabel="Age")

# Setup a mean line
ax0.axhline(y=over_50["chol"].mean(), 
            color='r', 
            linestyle='--', 
            label="Average");

ax0.legend(*scatter.legend_elements(), title="Target")


#################################################################################

# Axis 1, 1 (row 1, column 1)
scatter = ax1.scatter(over_50["age"], 
                      over_50["thalach"], 
                      c=over_50["target"],
                      cmap='winter')

ax1.set(title="Heart Disease and Max Heart Rate Levels",
        xlabel="Age",
        ylabel="Max Heart Rate")

# Setup a mean line
ax1.axhline(y=over_50["thalach"].mean(), 
            color='r', 
            linestyle='--', 
            label="Average");

ax1.legend(*scatter.legend_elements(), title="Target")

# Title the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```

_After the change of `xlim` and `ylim`_

```python
# Axis 0, 1 (row 0, column 1)
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 15))

scatter = ax0.scatter(over_50["age"], 
                      over_50["chol"], 
                      c=over_50["target"],
                      cmap='winter')

ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol",
        xlabel="Age")

# Setup a mean line
ax0.axhline(y=over_50["chol"].mean(), 
            color='r', 
            linestyle='--', 
            label="Average");

ax0.legend(*scatter.legend_elements(), title="Target")

# setting xlim and ylim for ax2
ax0.set_xlim([50,80])
ax0.set_ylim([100,600])

#################################################################################

# Axis 1, 1 (row 1, column 1)
scatter = ax1.scatter(over_50["age"], 
                      over_50["thalach"], 
                      c=over_50["target"],
                      cmap='winter')

ax1.set(title="Heart Disease and Max Heart Rate Levels",
        xlabel="Age",
        ylabel="Max Heart Rate")

# Setup a mean line
ax1.axhline(y=over_50["thalach"].mean(), 
            color='r', 
            linestyle='--', 
            label="Average");

ax1.legend(*scatter.legend_elements(), title="Target")

# setting xlim and ylim for ax1
ax1.set_xlim([50,80])
ax1.set_ylim([60,220])

# Title the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```
