## What is Data Engineering?


Well, Data Science is all about making businees decisions using data. One of ways that we can make businees decisions is to use a technique like Machine Learning. So, data science is an idea of leveraging data and converting it to make something useful for a product/business.

<br>

<br>

Data Analysis is a **subset** of Data Science that allows us to analyse data that we have, via certain automated tools/techniques.

<br><br>

In layman terms, a Data Engineer is responsible to collect data from all the sources (like cameras, IoT devices, website/app traffic, crash reports, etc.) and produces or maintain it in proper databases (structured/semi-structured/unstructured) and business can have access to this data in organised fashion.<br>
1. Structured - SQLite, PostgreSQL, etc.
2. Semi-Structured - XML, CSV, JSON, etc.<br>
3. Unstructured - PDFs, emails, documents, etc.
4. Binary - audio files, video (.mp4/.mp3/.mov), images.

<br><br>

### What Data Engineer does?

Mainly a data engineer creates "data pipelines". An abstract is shown below :
<br><br>

<br><br>

Basically what a **Data Engineer** does is they **build an ETL (Extract-Transform-Load) pipeline** - **e**xtract the data that's generated by devices at user's end and further organised in data lakes, **t**ransform it into a form that finally can be **l**oaded into a data warehouse. The languages used to accomplish these are, Scala, Go, Java, etc.
