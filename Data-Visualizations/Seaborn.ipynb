# ðŸ’¡ðŸ’¡ Simplified Cheatsheet to decide which plot to use when ðŸ’¡ðŸ’¡

To decide which plot to choose so as to demonstrate a relationship among different features/variables can be bit tricky sometimes. Well, this cheatsheet
can be helpful in that case.

__C__ - Categorical variable<br>
__N__ - Numerical variable

<hr>

__I.__ 1 variable i.e. Uni-variate analysis
- N => for only one Numerical variable
    1. histplot,
    2. KDEplot
    3. boxplot (results in a single bar)


- C => for only 1 Categorical variable
    1. countplot

<hr>

__II.__ 2 variable i.e. Bi-variate analysis
- N,N => two Numerical variables
    1. scatterplot
    2. jointplot with `hue`
    3. pairplot
    4. heatmap
    5. linechart


- C,C => two Categorical variables
    1. countplot with `hue`


- C,N => one Categorical + one Numerical variables
    1. boxplot
    2. violinplot
    3. barplot

<hr>

__III.__ 3 variable i.e. Tri-variate. But is useful in case of Multi-variate also.
- N,N,N => to graph relationship between three Numerical type variables/features
    1. scatterplot with `size` parameter


- C,C,C => show relationship between three Categorical variables - â›” (not possible, in general)


- N,C,C => one Numerical type + 2 Categorical type variables
    1. barplot with `hue`
    2. boxplot with `hue`
    3. violinplot with `hue`


- C,N,N => one Categorical type + two Numerical type vairbales
    1. scatterplot with `hue`
    2. jointplot with `hue`
    3. pairplots with `hue`
    4. linechart with `hue`

## `sns.kdeplot` - kernel density plot

- plots/depicts Probability Density of data i.e. distribution of entire values.
- The entire area under this plot = 1 (ie. probability = 1)
- If area under curve between two points - x1 & x2 - on x-axis is `A` . This denotes that our data may have values lying between x1-x2 with a probability of `A`.

```python
sns.kdeplot(iris["petal_width"])
```

- By looking at 2 peaks in above plot, we can decipher that approximately most petal width lengths are of either `0.3` or `1.5`.

```python
sns.kdeplot(iris["sepal_width"], shade=True)
```

- Similarly from above plot, approximately most sepal width are of length `3`.

- Area under entire curve is `1`.

## `sns.scatterplot`

```python
plt.figure(figsize=(10,5))

sns.scatterplot(data=iris,
                x="petal_length",
                y="petal_width")
```

Notice, the relationshiop between petal_width and petal_length which is - petal_width â¬† :: petal_length â¬†

> In above plot, can we inference which (petal_length, petal_width) coordinate belongs to which flower? -> __No__

So, let's introduce `hue` parameter :

```python
plt.figure(figsize=(8,5))

sns.scatterplot(data=iris,
                x="petal_length",
                y="petal_width",
                hue="species")
```

Points to remember :
- In above graph, 3 variables are involved.
- Use `scatterplot`, when 2 Continuous Numerical variables (NN) + 1 Categorical variable (C) (via `hue`) are involved.

If this same graph had to be plotted with Matplotlib, it would be done like this :

```python
setosa     = iris[iris["species"]=="setosa"]
virginica  = iris[iris["species"]=="virginica"]
versicolor = iris[iris["species"]=="versicolor"]

plt.scatter(x=setosa["petal_length"], y=setosa["petal_width"], c="r")
plt.scatter(x=virginica["petal_length"], y=virginica["petal_width"], c="g")
plt.scatter(x=versicolor["petal_length"], y=versicolor["petal_width"], c="b")

plt.show()
```

## `sns.jointplot`

```python
## histplot for petal_width + histplot for petal_length + scatter plot for petal_length & petal_width

sns.jointplot(data=iris,
               x="petal_length",
               y="petal_width")
```

```python
## kde for petal_width + kde for petal_length + scatter plot for petal_length & petal_width

sns.jointplot(data=iris,
               x="petal_length",
               y="petal_width",
               hue="species")
```

```python
j = sns.jointplot(data=iris,
                   x="petal_length",
                   y="petal_width",
                   hue="species",
                   kind="hist")

sns.move_legend(j.ax_joint, loc="upper left")
## move legend via firstly accessing the axes, then call .move_legend() on that axes

## the darker region implies, more number of flowers lie in that value region
```

```python
j = sns.jointplot(data=iris,
                   x="petal_length",
                   y="petal_width",
                   hue="species",
                   kind="kde")

sns.move_legend(j.ax_joint, loc="upper left")
## move legend via firstly accessing the axes, then call .move_legend() on that axes

## looks like Contour plot
```

## `sns.pairplot`

```python
## to depict relation between each of the pairs in -> sepal_length, sepal_width, petal_width, petal_length

sns.pairplot(data=iris,
             hue="species")
```

- The diagonal plots got chosen as Histogramic or KDE plots; because, such is a Uni-Variate scenario.
- Other plots are NNC type i.e. Numeric variable + Numeric variable + Categorical variable

## `sns.boxplot`

- Boxplot can be used to depict NC relationsip ie. Numerical (continuous) variable - Categorical variable.

```python
sns.boxplot(data=iris,
            x="species",         ## categorical
            y="sepal_width"     ## numerical
           )

plt.grid()
```

Few conclusion from above plot :
- Median(length of Setosa) > Median(length of Virginica) > Median(length of Versicolor)
- 75% values that belong to Versicolor (or, its 75 percentile) = Median length of Verginica <= 3
- min(length of Setosa) = Median(length of Virginica) = 75 percentile for Versicolot
- An Outlier point ("diamond" like points in the plot) for each flower is shown.
- For each flower, `IQR = 75 percentile value - 25 percentile value`. For eg, approximately IQR(for sepal-width of Setosa) = 3.8 - 3.3 = 0.5
- Whiskers are the minimum and maximum sepal-length values pertaining to each flower.

## `sns.violinplot`

```python
plt.figure(figsize=(10,5))

sns.violinplot(data=iris,
               x="species",
               y="sepal_width")

plt.grid()
```

Few conclusion from above plot :
- White dots are the Median values corresponding to each category/flower.
- This looks like a KDE + Boxplot.

## `sns.heatmap`

- ### .corr()

Correlation between different numerical type variables.

__Correlation Coefficient is a statistical measure of the strength of the relationship between the relative movements of 2 variables.__

1. negative correlation represents Inverse Relationship between those 2 variables.
2. Higher the |coeff| value, more is the correlation.
3. correlation = `1`; means, Complete Correlation. In fact, a variable would definitely be "Completely Correlated" to itself.


```python
iris.corr()
```

```python
sns.heatmap(iris.corr(),
            annot=True)

## Use `cmap` parameter to obtain heatmap of desired color.
```

```python
plt.figure(figsize=(10,6))

sns.heatmap(iris.corr(),
            annot=True,
            cmap="Blues",
            vmin=-1,
            vmax=1)
```

## `sns.barplot`

```python
sns.barplot(data=tips,
            x="day",
            y="total_bill")
```

__Intution :__
- barplot performs an aggregation (preferrably, "mean") on `y` parameter (here, `total_bill`) for each categorical value of `x`. This can be speciied via `estimator` parameter.
- black lines in each bar are related to Confidence Intervals for mean values w.r.t. each day. (as of now, out of scope topic!)
- Elaborately, on Thur the `mean("total_bill") ~ 17`. Similarly, `mean("total_bill") ~ 22` on Sunday.

```python
sns.barplot(data=tips,
            x="day",
            y="total_bill",
            estimator=np.min)
```

- The `min` total_bill value on Thursday was ~ 8 Rs.
- `min` total_bill for Sat happens to be lowest across all minimum values.
