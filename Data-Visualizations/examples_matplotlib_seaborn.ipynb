## <u>Q1.</u> Complete the program to interpret the Sales Revenue and Number-of-Customers of the website using two plots in the same figure.

The plot should follow the following conditions:

- The sales over the years should be plotted using squares.
- The number of customers should be plotted using a green line in the same graph.
- The y-axis label should be `sales & number of customers`.
- The x-axis label should be `years`.
- The legends should be `Sales (in millions)` and `Number of customers (in thousands)`.
- The title of plot should be 'Analysis of Sales and Customers'.

```python
year            = [2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013]
sales           = [8457, 2813, 1061, 5341, 9308, 9372, 12383, 15040, 20060, 18585]
no_of_customers = [12478, 942, 1227, 2660, 8832, 7703, 10758, 15398, 16934, 14257]

df = pd.DataFrame({'year': year,
                   'sales': sales,
                   'number_of_customers': no_of_customers})


fig = plt.figure(figsize=(8,8))

#Create the plot sales and years
# Or, can be scatter plot also -> plt.scatter(x, y, marker="s", label="Sales(in millions)")
plt.plot(df["year"], df["sales"], "s", label="Sales (in millions)")


#Create the plot number_of_customers and years
plt.plot(df["year"], df["number_of_customers"], color="red", label="Number of customers (in thousands)")

#Add the xlabel
plt.xlabel("years")

#Add the ylabel
plt.ylabel("sales & number of customers")

#Add the legends and grid
plt.legend()
plt.grid()

#Add text for 2008's Target
plt.text(2008, 6000, s="- 2008's Sales Target of 6K $")

#Add the title
plt.title('Analysis of sales and customers')
```

## <u>Q5.</u> Given the data "age of biker" and "biker's Avg laptime", create a scatter plot using squares ('s') for representing points, in which values on the x-axis and y-axis are limited by `x_lim` and `y_lim`.

```python
age        = [25, 24, 35, 24, 29, 28, 29, 38, 28, 21]
avgLaptime = [8.9, 14.3, 12.4, 9.1, 10.3, 9.8, 11.8, 13.5, 8.0, 13.3]


fig = plt.figure(figsize=(8,5))

#plot the points using 's' with default color
plt.scatter(x=age,
            y=avgLaptime,
            marker="s")

x_ticks = np.arange(0, 45, 2)
#Allot the ticks to the x-axis
plt.xticks(x_ticks)

y_ticks = np.arange(5, 15, 0.5)
#Allot the ticks to the y-axis
plt.yticks(y_ticks)

x_lim = (15, 45)
#Limit the values on x-axis according to x_lim
plt.xlim(x_lim)

y_lim = (5, 15)
#Limit the values on y-axis according to y_lim
plt.ylim(y_lim)

#Add the text "trial" to point (age[0], avgLaptime[0])
plt.text(x=age[0],
         y=avgLaptime[0],
         s="trial")

plt.grid()
plt.xlabel("Age")
plt.ylabel("Avg Laptime")
plt.title("Biker's Age v/s His Avg Laptime")
plt.show()
```
