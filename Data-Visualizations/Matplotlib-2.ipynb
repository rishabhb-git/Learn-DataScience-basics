### Bar charts are useful for visualizing counts, or summary statistics with error bars.

### This example shows a way to create a "grouped bar chart" (or, "Dodged" bar chart) with Matplotlib and also how to annotate bars with labels.

ref : https://matplotlib.org/3.1.0/gallery/lines_bars_and_markers/barchart.html

```python
men_means, men_std = (20, 35, 30, 35, 27), (2, 3, 4, 1, 2)
women_means, women_std = (25, 32, 34, 20, 25), (3, 5, 2, 3, 3)

ind = np.arange(len(men_means))  # the x locations for the groups
width = 0.35                     # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/2, men_means, width, yerr=men_std,
                label='Men')
rects2 = ax.bar(ind + width/2, women_means, width, yerr=women_std,
                label='Women')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(ind)
ax.set_xticklabels(('G1', 'G2', 'G3', 'G4', 'G5'))
ax.legend()


def autolabel(rects, xpos='center'):
    """
    Attach a text label above each bar in *rects*, displaying its height.

    *xpos* indicates which side to place the text w.r.t. the center of
    the bar. It can be one of the following {'center', 'right', 'left'}.
    """

    ha = {'center': 'center', 'right': 'left', 'left': 'right'}
    offset = {'center': 0, 'right': 1, 'left': -1}

    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(offset[xpos]*3, 3),  # use 3 points offset
                    textcoords="offset points",  # in both directions
                    ha=ha[xpos], va='bottom')


autolabel(rects1, "left")
autolabel(rects2, "right")

fig.tight_layout()

plt.show()
```

## `plt.hist()`

- returns a tuple, alonside plotting a Histogram.

- `bins` = 10, by default

```python
values = [2, 1, 3, 2, 1, 0, 4, 3, 1, 4, 0, 3, 1]

plt.hist(values)
```

Above graph shows a Frequency Distribution , i.e. :
- frequency of each number in the `values` on y-axis.
- a number in `values` list , on x-axis.

Below histogram shows frequency within the ranges. This is achieved via, `bins` parameter.

Meaning, below plot involve 4 bins. The x-axis ticks for these bins are created, by taking the least number (= `0`) and the highest number (= `4`)  that occur in `num_list` list, divided by the `bins` parameter.
- So here, bins are :  `[0-1)`, `[1-2)`, `[2-3)` and finally `[3-4]`.

Count of all values that fall inside the `[0, 1)` bin - i.e. 0 inluded and excluded - is shown on y-axis. SImilarly, for other bins.

```python
num_list = [2, 1, 3, 2, 1, 0, 4, 3, 1, 4, 0, 3, 1]

plt.hist(num_list, bins=4)
```

<br>

## `plt.scatter()`

```python
a = np.arange(0,20)
b = np.linspace(1, 20, dtype=np.float32, num=20) + np.random.rand(20) * 10

plt.scatter(x= a,
            y= b)

print(a)
print("-" * 20)
print(b)
plt.show()
```

In above example, there are essentially 2 variables involved; for example, x=Date, y=Sales (in INR.)

<br>

However in below example, there are 3 variables involved - x=Year, y=Population, third-variable=Size.

```python
year = [1950, 1970, 1990, 2010]
population = {"india" : [30, 52, 85, 108],
              "USA"   : [40, 55, 70, 90],
              "China" : [60, 80, 90, 120]
             }

plt.scatter(x= year,
            y= population["india"],
            c= "b",
            s= population["india"])           ## `s` is Size of dots

plt.scatter(x= year,
            y= population["USA"],
            c= "r",
            s= population["USA"])

plt.scatter(x= year,
            y= population["China"],
            c= "g",
            s= population["China"])

plt.legend(["India", "USA", "China"])
plt.grid()
plt.xlabel("Year")
plt.ylabel("Pupulation (in crores)")
plt.title("Populations v/s Year for India, USA, China")
plt.show()
```

## `plt.subplots()`

```python
x = np.arange(-10, 10, 0.1)
y1 = np.sin(x)
y2 = x ** 2

fig, ax = plt.subplots(2,1, figsize=(10, 5))

ax[0].plot(x, y1)
ax[1].plot(x, y2)

ax[0].set(xlabel="x ->", ylabel="sin of x")
ax[0].set_title("y = sin(x)")

ax[1].set(xlabel="x ->", ylabel="square of x")
ax[1].set_title("y = x²")
```

```python
x = np.arange(-5, 5, 0.1)
y1 = np.sin(x)
y2 = x ** 2
y3 = abs(x) + 1
y4 = np.cos(x)

fig, ax = plt.subplots(2,2, figsize=(8, 10))

## plotting all subplots
ax[0,0].plot(x, y1)
ax[0,1].plot(x, y2)
ax[1,0].plot(x, y3)
ax[1,1].plot(x, y4)

## for sin(x)
ax[0,0].set(xlabel="x ->", ylabel="sin of x")
ax[0,0].set_title("y = sin(x)")

## for x**2
ax[0,1].set(xlabel="x ->", ylabel="square of x")
ax[0,1].set_title("y = x²")

## for abs(x)+1
ax[1,0].set(xlabel="x ->", ylabel="|x| + 1")
ax[1,0].set_title("y = |x| + 1")

## for coz(x)
ax[1,0].set(xlabel="x ->", ylabel="cos(x)")
ax[1,0].set_title("y = cos(x)")
```

## 3-D Graphs

- Matplotlib allows us to plot 3-D graphs

```python
a = np.array([0, 1, 2, 3])
b = np.array([0, 1, 2])

a, b = np.meshgrid(a, b)
```

```python
a
```

```python
b
```

#### Notice here:
- `meshgrid()` - allows you to **convert a 1-D array into 2-D**

#### How is `meshgrid()` working?

- It **takes dimensions of both arrays `a` and `b`**
    - Dimension of `a` = (4,)
    - Dimension of `b` = (3,)


- **Creates two `3 x 4` matrices**


- In **first matrix**, array of **`a` repeats 3 times** and gets **stacked vertically**


- In **second matrix**, array of **`b` repeats 4 times** and gets **stacked horizontally**


#### So, we get all possible combinations as coordinates

`
(0,0)  (1, 0)  (2, 0)  (3, 0)
(0,1)  (1, 1)  (2, 1)  (3, 1)
(0,2)  (1, 2)  (2, 2)  (3, 2)
`


- **Two 1-D arrays get converted into a meshgrid**


#### We can use the meshgrid to plot our 3-D graph

```python
fig = plt.figure(figsize=(8,8))

ax = fig.add_subplot(projection='3d')
ax.plot_surface(a, b, np.array([[7]]) )  # 3rd argument is the elevation in 3-D

plt.show()
```

#### Observe that:

- **Range on the axes of our plane surface** is taken from **`a` and `b`**

    - **x-axis is represented by `a`** ---> 0 to 3
    
    - **y-axis is represented by `b`** ---> 0 to 2    


- The **3rd dimension is the height from the floor we passed in**

    - **z-axis is represented by the elevation** we provided ---> 7

<br>

### Let's build a more complex 3-D plot

```python
a = np.arange(-1, 1, 0.005)
b = a  
a, b = np.meshgrid(a, b)


# Since `a` and `b` are some, this time we'll get a square matrix
```

```python
a.shape
```

```python
b.shape
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.gca(projection='3d')
ax.plot_surface(a, b, a**2 + b**2) # We'll NOT keep a constant z-axis elevation this time
plt.show()
```

#### Notice that:

- **Higher the values of `a` and `b`, higher will be z-axis value**


- At **`(a, b) = (0, 0)`**, `z` is **lowest at 0**


- The **dimensions of `a`, `b` and `a^2 + b^2` are all same** = $400\times400$
    

- That is why it is able to map all **(x, y, z) coordinates.**

## Pie charts

Finally Matplotlib also provide you with **Pie charts.**

You can use the following basic syntax to create a pie chart from a pandas DataFrame:

`df.groupby(['group_column']).sum().plot(kind='pie', y='value_column')`

The following examples show how to use this syntax in practice.

We can use the following syntax to create a pie chart that displays the portion of total points scored by each team:

```python
plt.pie([10,20,30,40],
       labels=['Railways', 'Education', 'Agriculture', 'Defence'],
       colors=['c','m','r','b'],
       startangle=90,
       shadow=True,
       explode=(0, 0.15, 0.05, 0))

plt.show()
```
